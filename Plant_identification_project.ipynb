```python
from duckduckgo_search import ddg_images
from fastcore.all import *

def search_images(term, max_images=200): return L(ddg_images(term, max_results=max_images)).itemgot('image')
```

```python
#results = search_images_bing(key, 'grizzly bear')
#adapted to use ddg instead
results = search_images('turkey oak leaf', max_images=150)
ims = results.attrgot('contentUrl')
len(ims)
```

```python
oak_types = 'petraea','robur'
path = Path('oaks')
```

```python
if not path.exists():
    path.mkdir()
for o in oak_types:
    dest = (path/o)
    dest.mkdir(exist_ok=True)
    #results = search_images_bing(key, f'{o} bear')
    results = search_images(f'{o} oak', max_images=150)
    content = results
    print(results[0:10])
    print(o, len(content))
    download_images(dest, urls=content)
```

```python
!find oaks
```

```python
fns = get_image_files(path)
fns
```

```python
failed = verify_images(fns)
failed
```

```python
failed.map(Path.unlink);
```

```python
oaks = DataBlock(
    blocks=(ImageBlock, CategoryBlock),
    get_items=get_image_files,
    splitter=RandomSplitter(valid_pct=0.2, seed=42),
    get_y=parent_label,
    item_tfms=Resize(128))
```

```python
oaks
```

```python
dls = oaks.dataloaders(path)
```

```python
dls.valid.show_batch(max_n=4, nrows=1)
```

```python
oaks = oaks.new(
    item_tfms=RandomResizedCrop(224, min_scale=0.5),
    batch_tfms=aug_transforms())
dls = oaks.dataloaders(path)
```

```python
learn = vision_learner(dls, resnet18, metrics=error_rate)
learn.fine_tune(4)
```

```python
interp = ClassificationInterpretation.from_learner(learn)
interp.plot_confusion_matrix()
```

```python
interp.plot_top_losses(5, nrows=1)
```

```python
#hide_output
cleaner = ImageClassifierCleaner(learn)
cleaner
```
